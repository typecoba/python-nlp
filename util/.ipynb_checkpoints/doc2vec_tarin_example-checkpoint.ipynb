### preprocessing

```python
%time
import os
import pickle
import csv
import pandas as pd
import random
from gensim.models import Doc2Vec
from gensim.models.doc2vec import TaggedDocument
```

```python
# load
class HashtagDoc2Vec:
    def __init__(self, fname):
        self.fname = fname
        if not os.path.exists(fname):
            print('File not found: %s' % fname)
    
    def __iter__(self):
        with open(self.fname, 'r', encoding='utf-8') as f:                    
            for row in f :    
                ids, tags = row.split('\t')
                tagged_doc = TaggedDocument(
                    words = tags.split(),
                    tags = [ids]
                )
                yield tagged_doc
```

```python
openPath = '../data/ta_hashtag/doc2vec_ta_hashtag_filtered.txt'
doc2vec_corpus = HashtagDoc2Vec(openPath)
```

```python
 for i, doc in enumerate(doc2vec_corpus):
    if i > 3: break
    print(doc)
```

### training

word2vec과 마찬가지로 parameter 조정 필요...

```python
%%time
TRAIN_DOC2VEC = False
doc2vec_fname = '../data/ta_hashtag/hashtag_doc2vec_model_filtered.pkl'

if TRAIN_DOC2VEC : 
    doc2vec_model = Doc2Vec(
        doc2vec_corpus,
        dm=1,
        vector_size=200,
        alpha=0.015,
        window=6,
        min_count=10,
        workers=8,
        negative=5
    )
    with open(doc2vec_fname, 'wb') as f:
        pickle.dump(doc2vec_model, f)
else :
    with open(doc2vec_fname, 'rb') as f:
        doc2vec_model = pickle.load(f)
```

```python
# pickle save
loadPath = '../data/ta_hashtag/doc2vec_ta_hashtag_idx2name.txt'
with open(loadPath, 'r') as f :    
    idx2name={}
    for num, doc in enumerate(f):
        temp = doc.split('\t')
        idx2name[temp[0]] = {'username':temp[1], 'follower_cnt':int(temp[2]), 'following_cnt':int(temp[3]), 'media_cnt':int(temp[4].replace('\n',''))}
```

```python
savePath = '../data/ta_hashtag/doc2vec_ta_hashtag_idx2name.pkl'
with open(savePath, 'wb') as f:
    pickle.dump(idx2name, f)
f.close()    
```

```python
# load idx
loadPath = '../data/ta_hashtag/doc2vec_ta_hashtag_idx2name.pkl'
with open(loadPath, 'rb') as f:
    idx2name = pickle.load(f)
f.close()    
```

```python
'''
fit data
doc2vec_model - idx2vec
dataframe으로 묶어서 처리
''' 
fit_data = []
for num, row in enumerate(doc2vec_model.docvecs.doctags) : 
#     if num > 1 : break    
    temp_data = idx2name[row]
    temp_data['id'] = row
    temp_data['vector'] = doc2vec_model.docvecs.doctag_syn0[num]
    fit_data.append( temp_data )
```

```python
fit_df = pd.DataFrame(fit_data)
```

### bokeh

```python
from bokeh.models.callbacks import CustomJS
from bokeh.models import TapTool, HoverTool, ColumnDataSource, OpenURL
from bokeh.plotting import figure, show, output_notebook
output_notebook()    
```

### scatter plot

```python
def scatterplot_similar_idx(labels, embeddings):
    # tools
    hover = HoverTool(names=['circle'],tooltips=[('id', '@id'),
                                                ('username','@username'),
                                                ('media_cnt','@media_cnt'),
                                                ('follower_cnt','@follower_cnt'),
                                                ('following_cnt','@following_cnt')])
    
#     taptool = TapTool(callback=CustomJS(code="window.open('https://www.instagram.com/"+"@username"+"','_blank');wnidoe.focus();"))
    taptool = TapTool(callback = OpenURL(url='https://www.instagram.com/@username'))
    
    tools = [hover, taptool, 'pan','wheel_zoom','reset']
    
    # set figure
    p = figure(title='t-sne', toolbar_location='left', tools=tools)
    p.grid.grid_line_color = None
    p.background_fill_color = "white"
    p.width = 900
    p.height = 900

    
    # circle    
    source = ColumnDataSource( 
        data = dict(
            x = embeddings[:,0],
            y = embeddings[:,1],
            id = sample_df['id'],
            username = sample_df['username'],
            media_cnt = sample_df['media_cnt'],
            follower_cnt = sample_df['follower_cnt'],
            following_cnt = sample_df['following_cnt']
    ))
    p.circle('x', 'y', name="circle", source=source,
             size=5, line_color="navy", fill_color="orange", alpha=0.5)    
    

    
    # show
    show(p)

scatterplot_similar_idx(idx2name, y_tsne)
```

```python
def get_most_similar(username,count):
    
    idx = sample_df[sample_df['username']==username]['id'].values[0]
    
    for row in doc2vec_model.docvecs.most_similar(str(idx), topn=count) :         
        print( "https://www.instagram.com/{}\t\t{}".format(idx2name[row[0]]['username'],row[1]) )
```
