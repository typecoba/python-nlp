```python
from soynlp.utils import DoublespaceLineCorpus
from pprint import pprint
fpath = './data/hashtag/md_cpm_hashtag.txt'
```

```python
corpus = DoublespaceLineCorpus(fpath, iter_sent=True)
```

```python
pprint(list(corpus)[:50])
```

```python
tags = []
for i,sent in enumerate(corpus) :     
    splited_sent = sent.split(',')
    for word in splited_sent : 
        if word not in tags : 
            tags.append(word)            
```

```python
print(len(tags))
pprint(tags)
```

```python
# 저장
data ='' 
with open('./data/hashtag/md_cpm_hashtag_preprocessed.txt', 'w', encoding='utf-8') as f :
    for word in tags : 
        data += word+'\n'
    f.write(data)
    f.close
```

```python
# tag analyis에 있는지 확인
import requests
import json
api = 'https://b8k9yk75eb.execute-api.ap-northeast-2.amazonaws.com/tag_api/tag-analysis/tag-similar-request-by-name'
```

```python
temp = []
for i, word in enumerate(tags) : 
#     if i > 10 : break
    try :
        params = {'p':word}
        request_data = requests.get(api, params=params).text
        json_data = json.loads(request_data)
        if 'errorType' in json_data :
            temp.append(word)
            print(i, len(temp), word)
    except Exception as ex : 
        print(ex)
```
